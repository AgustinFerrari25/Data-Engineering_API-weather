# 2. Connect to the api and extract the data

```python
import requests

url = f'https://api.weatherapi.com/v1/forecast.json?key=93b9f367467b459393a225415241002&q=Buenos Aires&days=4&aqi=no&alerts=no'
response = requests.get(url)
data = response.json()

forecast_data_dict = {}

# Iterar sobre los datos de pronóstico
for forecast_day in data.get('forecast', {}).get('forecastday', []):
    hourly_data = forecast_day.get('hour', [])
    date = forecast_day.get('date')
    
    # Iterar sobre los datos de pronóstico cada 2 horas
    for i in range(0, len(hourly_data), 2):
        hour_data = hourly_data[i]
        time = hour_data.get('time')
        forecast_data_dict[time] = {
            'Temperature_C': hour_data.get('temp_c'),
            'Condition': hour_data.get('condition', {}).get('text'),
            'Wind Speed_kph': hour_data.get('wind_kph'),
            'wind Direction': hour_data.get('wind_dir'),
            'Pressure_mb': hour_data.get('pressure_mb'),
            'Humidity%': hour_data.get('humidity'),
            'Cloud_Cover%': hour_data.get('cloud'),
            'chance_of_rain%': hour_data.get('chance_of_rain'),
            'Precipitation_mm': hour_data.get('precip_mm'),
            'UV_Index': hour_data.get('uv'),
        }

forecast_data_dict

```

```python
# Crear DataFrame de Pandas
import pandas as pd
from datetime import datetime

df = pd.DataFrame.from_dict(forecast_data_dict, orient='index').reset_index().rename(columns={"index": 'Hour'})

creation_date = datetime.now()
df['Creation_Date'] = creation_date

df
```

```python
import psycopg2
from dotenv import load_dotenv
import os

#cargamos las variables del archivo .env
load_dotenv()

host_instance = os.getenv("AWS_HOST")
data_base = os.getenv("AWS_DB")
user = os.getenv("AWS_USER")
port=os.getenv('AWS_PORT')
pwd = os.getenv("AWS_PASSWORD")

try:
    conn = psycopg2.connect(
        host=host_instance,
        dbname=data_base,
        user=user,
        password=pwd,
        port=port
    )
    print("successful connection to AWS Redshift")
except Exception as e:
    print("error connecting to AWS Redshift")
    print(e)
```

```python
def cargar_en_redshift(data_base_name, conn, df):
    try:
        cur = conn.cursor()

        # Verificar si la tabla ya existe
        cur.execute(f"SELECT 1 FROM information_schema.tables WHERE table_name = '{data_base_name}'")
        if cur.fetchone():
            print("Table already exists")
        
        else:
    
        # Tabla en Redshift
            dtypes = df.dtypes
            cols = list(dtypes.index)
            tipos = list(dtypes.values)
    
            type_map = {'int64': 'INTEGER', 'float64': 'FLOAT', 'object': 'VARCHAR(50)', 'datetime64[us]':'TIMESTAMP'}
    
            sql_dtypes = [type_map[str(dtype)] for dtype in tipos]
    
            column_defs = [f'"{name}" {data_type}' for name, data_type in zip(cols, sql_dtypes)]
    
            tabla = f"CREATE TABLE IF NOT EXISTS {data_base_name} ({', '.join(column_defs)});"
    
            cur.execute(tabla)
            print("Table created successfully")
    
        # Insertar datos en la tabla
        for index, row in df.iterrows():
            values = [str(value) if isinstance(value, pd.Timestamp) else value for value in row]
            insert_query = f"INSERT INTO {data_base_name} VALUES ({', '.join(['%s'] * len(values))});"
            cur.execute(insert_query, values)
    
        conn.commit()
        print("Data inserted successfully")
        #verificamos si hay valores repetidos en la columna de 'Hour'
        cur.execute(f"SELECT COUNT(*) FROM (SELECT Hour FROM {data_base_name} GROUP BY Hour HAVING COUNT(*) > 1) AS duplicates;")
        result = cur.fetchone()
        num_duplicates = result[0]

        if num_duplicates > 0:
            #si hay duplicados, se elimina las filas duplicadas mas nuevas
            cur.execute(f"""DELETE FROM {data_base_name} WHERE EXISTS (SELECT 1 FROM (SELECT Hour, MIN(Creation_Date) AS min_creation_date FROM {data_base_name} GROUP BY Hour HAVING COUNT(*) > 1 ) AS dup WHERE {data_base_name}.Hour = dup.Hour AND {data_base_name}.Creation_Date > dup.min_creation_date);""")
            
            conn.commit()
            print("Duplicate rows deleted successfully")
        else:
            print("No duplicate rows found, skipping deletion.")

    except Exception as e:
        print("Error inserting data")
        print(e)

    finally:
        cur.close()
        conn.close()
```
